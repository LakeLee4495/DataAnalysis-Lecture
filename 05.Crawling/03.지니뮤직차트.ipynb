```python
import requests
import pandas as pd
```

```python
url='https://www.genie.co.kr/chart/top200'
req = requests.get(url)
html = req.text
html
```

```python
header = {'User-Agent':
          'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.88 Mobile Safari/537.36'}
req = requests.get(url, headers=header)
html = req.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, 'html.parser')
```

```python
table = soup.select_one('table.list-wrap')
trs = table.select('tr.list')
len(trs)
```

```python
tr= trs[0]
tr.select_one('.number').get_text()
```

```python
tr.select_one('.number').get_text().split('\n')[0]
```

```python
rank= int(tr.select_one('.number').get_text().split('\n')[0])
rank
```

```python
title= tr.select_one('.info').select_one('.title').get_text().strip()
title
```

```python
artist= tr.select_one('.info').select_one('.artist').get_text().strip()
artist
```

```python
album= tr.select_one('.info').select_one('.albumtitle').get_text().strip()
album
```

```python
rank_list, title_list, artist_list, album_list = [], [],[],[]

for tr in trs:
  rank= int(tr.select_one('.number').get_text().split('\n')[0])
  title= tr.select_one('.info').select_one('.title').get_text().strip()
  artist= tr.select_one('.info').select_one('.artist').get_text().strip()
  album= tr.select_one('.info').select_one('.albumtitle').get_text().strip()
  rank_list.append(rank)
  title_list.append(title)
  artist_list.append(artist)
  album_list.append(album)
```

```python
df = pd.DataFrame({
    '순위': rank_list,
    '제목' : title_list,
    '가수' : artist_list,
    '앨범명' : album_list
})
df
```

```python
df.to_csv('지니차트(20220414)Top50.csv')
```
