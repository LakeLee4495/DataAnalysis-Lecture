- 과제 \
10군데 핫 플레이스 \
지도 위에 표시(툴팁: 장소이름, 팝업: 특정 메세지) \
아이콘 모양을 전부 다르게


```python
import folium
```

```python
map = folium.Map(location=[36, 127], zoom_start=8)
folium.Marker(
    location=[37.5164587, 127.072772],                                          # 잠실
    icon = folium.Icon(color='red', icon='star'),
    popup=folium.Popup('LG트윈스', max_width=200),
    tooltip='서울종합운동장 야구장'
).add_to(map)
folium.Marker(
    location=[37.5164587, 127.072972],                                          # 잠실
    icon = folium.Icon(color='black', icon='flag'),
    popup=folium.Popup('두산베어스', max_width=200),
    tooltip='서울종합운동장 야구장'
).add_to(map)
folium.Marker(
    location=[37.4996453, 126.866921],                                          # 고척
    icon = folium.Icon(color='purple', icon='home'),
    popup=folium.Popup('키움히어로즈', max_width=200),
    tooltip='고척 스카이돔'
).add_to(map)
folium.Marker(
    location=[37.4360004, 126.690913],                                          # 인천
    icon = folium.Icon(color='red', icon='off'),
    popup=folium.Popup('SSG 랜더스', max_width=200),
    tooltip='SSG 랜더스필드'
).add_to(map)
folium.Marker(
    location=[35.1684217, 126.888301],                                          # 광주
    icon = folium.Icon(color='red', icon='minus'),
    popup=folium.Popup('KIA 타이거즈', max_width=200),
    tooltip='광주-기아 챔피언스 필드'
).add_to(map)
folium.Marker(
    location=[35.8410990, 128.681260],                                          # 대구
    icon = folium.Icon(color='blue', icon='plus'),
    popup=folium.Popup('삼성 라이온즈', max_width=200),
    tooltip='대구 삼성 라이온즈 파크'
).add_to(map)
folium.Marker(
    location=[36.3165578, 127.430646],                                          # 대전
    icon = folium.Icon(color='orange', icon='ok'),
    popup=folium.Popup('한화 이글스', max_width=200),
    tooltip='대전 한화생명 이글스파크'
).add_to(map)
folium.Marker(
    location=[35.1934884, 129.064317],                                          # 부산
    icon = folium.Icon(color='beige', icon='cog'),
    popup=folium.Popup('롯데 자이언츠', max_width=200),
    tooltip='사직 야구장'
).add_to(map)
folium.Marker(
    location=[37.2958751, 127.015086],                                          # 수원
    icon = folium.Icon(color='black', icon='cloud'),
    popup=folium.Popup('KT WIZ', max_width=200),
    tooltip='수원 케이티 위즈 파크'
).add_to(map)
folium.Marker(
    location=[35.2216105, 128.579193],                                          # 마산
    icon = folium.Icon(color='lightblue', icon='user'),
    popup=folium.Popup('NC 다이노스', max_width=200),
    tooltip='창원 NC 파크'
).add_to(map)
title_html = '<h3 align="center" style="font-size:20px">10개 국내 1군 야구장 표시</h3>'    
map.get_root().html.add_child(folium.Element(title_html)) 
map
```
