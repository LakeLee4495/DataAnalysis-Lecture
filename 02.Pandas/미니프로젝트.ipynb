## cctv 데이터 정리하기
- 데이터 파일 읽기
- Null data 있는지 확인하기
- 1000 단위 구분기호 없애고 정수로 변환하기
- 2021년을 제외한 최근 3개년간의 CCTV 증가율 구하기
- 구분, 총계, 최근증가율 컬럼만 추출하고 컬럼명을 구별, CCTV댓수로 변경하기
- 구 이름에서 공백지우기
- 첫번째 행(계) 지우기

## 인구 데이터 정리하기
 - 데이터 파일 읽기
 - Null data 있는지 확인하기
 - '자치구','계','계.1','계.2','65세 이상 고령자' 컬럼만 추출하고
   이름을 '구별', '인구수','내국인','외국인','고령자'로 변경
 - 1000단위 구분기호 없애고 정수로 변환하기
 - 첫번째 행 지우기
 - '외국인 비율'(외국인/인구수)과 '고령자 비율'(고령자/인구수) 컬럼 만들기

## cctv 데이터와 인구 데이터 합치고 저장하기

- 데이터 합치기
- 구별 컬럼을 인덱스로 만들기
- 합친 데이터를 cctv_pop.csv 로 저장하기

```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# 데이터 파일 읽기
filename = '서울시CCTV설치운영현황(자치구)_년도별_211231기준.csv'
cctv = pd.read_csv(filename, skiprows=1, encoding='EUC-KR')
cctv
```

```python
# null 값 여부 확인
cctv.isna().sum()
```

```python
# null 값을 0으로 대체
cctv.fillna(0, inplace=True)
cctv
```

```python
# 1000 단위 구분기호 없애고 정수로 변환하기 (불러올때 정수로 불러옴)
cctv = pd.read_csv(filename, skiprows=1, encoding='EUC-KR',thousands=',')
cctv.fillna(0, inplace=True)
cctv
```

```python
# CCTV 데이터 정리하기
cctv['2017년_이전'] = cctv.총계 - cctv['2018년'] - cctv['2019년'] - cctv['2020년'] - cctv['2021년']
cctv['최근증가율'] = ((cctv['2018년']+cctv['2019년']+cctv['2020년']) / cctv['2017년_이전'] * 100).round(2)
```

```python
# 구분, 총계,최근증가율 컬럼만 추출
cctv = cctv[['구분','총계','최근증가율']]
```

```python
# 구분,총계 명칭 변경
cctv.rename(columns={'구분':'구별','총계':'CCTV댓수'}, inplace=True)
cctv
```

```python
# 공백 제거
cctv['구별'] = cctv.구별.apply(lambda x: x.replace(' ',''))
cctv
```

```python
#첫번째 행 지우기 (계값 없어짐)
cctv.drop([0], inplace=True)
cctv
```

```python
# 데이터 파일 읽기 (1000단위 구분기호 없애고 정수로 변환 작업 병행)
filename = '서울시인구.txt'
pop = pd.read_csv(filename, sep='\t', skiprows=2, thousands=',')
pop
```

```python
# Null data 있는지 확인
pop.isna().sum()
```

```python
# 자치구','계','계.1','계.2','65세 이상 고령자' 컬럼만 추출하고 이름을 '구별', '인구수','내국인','외국인','고령자'로 변경
pop = pop[['자치구','계','계.1','계.2','65세이상고령자']]
pop
```

```python
pop.rename(columns={'자치구':'구별','계':'인구수','계.1':'내국인','계.2':'외국인','65세이상고령자':'고령자'}, inplace=True)
pop
```

```python
# 1000 단위 구분기호 없애고 정수로 변환하기 (불러올때 이미 설정함 thousands=',')
```

```python
# 첫번째 행 지우기 (합계행 없어짐)
pop.drop([0], inplace=True)
pop
```

```python
# 외국인 비율과 고령자 비율 컬럼 만들기
pop['외국인비율'] = (pop.외국인/pop.인구수*100).round(2)
pop['고령자비율'] = (pop.고령자/pop.인구수*100).round(2)
pop
```

- 데이터 합치기

```python
# data 합치기
df = pd.merge(cctv, pop)
df
```

```python
# '구별' 컬럼을 인덱스로 만들기
df.set_index('구별', inplace=True)
df
```

```python
# 합친 데이터를 cctv_pop.csv로 저장하기
df.to_csv("cctv_pop.csv", mode='w')
```
